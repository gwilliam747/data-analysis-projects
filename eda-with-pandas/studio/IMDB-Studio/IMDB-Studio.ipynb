### 5. Data Selection - Based on Conditional Filtering

```python
# Use this code block to select movies that have been released between 2010-2016, with a rating of less than 6.0, but generated significant revenue.
# For "significant revenue", filter for movies in the top 25% of revenue (use data['Revenue (Millions)'].quantile(0.75))

selected_movies = imdb_indexed[(imdb_indexed["Year"] >= 2010)
                               & (imdb_indexed["Year"] <= 2016) 
                               & (imdb_indexed["Rating"] < 6.0) 
                               & (imdb_indexed["Revenue (Millions)"] >= imdb_indexed["Revenue (Millions)"].quantile(0.75))]
print(selected_movies)
```

```python
# Bonus Challenge: Try adjusting the quantile value to be more specific.
selected_movies = imdb_indexed[(imdb_indexed["Year"] >= 2010)
                               & (imdb_indexed["Year"] <= 2016) 
                               & (imdb_indexed["Rating"] < 6.0) 
                               & (imdb_indexed["Revenue (Millions)"] >= imdb_indexed["Revenue (Millions)"].quantile(0.90))]
#print(selected_movies)
# Can you find movies with exactly the characteristics you're looking for by experimenting with different thresholds?
selected_movies = imdb_indexed[(imdb_indexed["Year"] >= 2010)
                               & (imdb_indexed["Year"] <= 2016) 
                               & (imdb_indexed["Rating"] < 5.0) 
                               & (imdb_indexed["Revenue (Millions)"] >= imdb_indexed["Revenue (Millions)"].quantile(0.90))]
print(selected_movies)
# Two twilight movies take the cake for worst high-grossing bad movies released 2010-2016
# They are the only ones to rank in the 90th percentile for revenue with a rating of less than 5/10
```
