## Object Creation

Before coding, think of three things you want to make. These could be coding projects, new recipes, or planned projects for a woodworking hobby. Brainstorm the appropriate column names and data types as well.

Create a table under your schema called `planned_makes`.

```python
CREATE TABLE Sep2025JunkDB.henry_geerling.planned_makes (
    recipe_id INT IDENTITY(1, 1) PRIMARY KEY,
    recipe_name VARCHAR(255),
    meal VARCHAR(255),
    dish_type VARCHAR(255),
    uses_grill BIT,
    uses_stove BIT,
    uses_oven BIT,
    uses_blender BIT,
    uses_egg_beater BIT,
);
```

Add three rows to your table that are your planned makes.

```python
INSERT INTO Sep2025JunkDB.henry_geerling.planned_makes
    (recipe_name, meal, dish_type, uses_grill, uses_stove, uses_oven, uses_blender, uses_egg_beater)
VALUES ('Grilled Chicken', 'Dinner', 'Entree', '1', '0', '0', '0', '0'),
    ('Bacon and Hash Brown Omlette', 'Breakfast', 'Entree', '0', '1', '0', '0', '0'),
    ('Chocolate Chip Cookies', 'Dessert', 'appetizer', '0', '0', '1', '0', '1');
```

## Update a Row

Change different values in one of your three rows.

```python
UPDATE Sep2025JunkDB.henry_geerling.planned_makes
SET dish_type = 'Appetizer'
WHERE dish_type = 'appetizer'
```
