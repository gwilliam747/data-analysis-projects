### Event 1: Women's History Month

Highlight popular women writers based on ratings from `BooksDB` by writing a query that returns `tag_id`, the number of times each `tag_id` is used and the `tag_name`. Use the `GROUP BY` and `HAVING` clause to narrow your focus and try multiple keywords, such as "woman" and "female".

```python
-- Solution
SELECT bks.title, bt.goodreads_book_id, bt.tag_id, bt.count, bks.book_id, bks.average_rating, tgs.tag_id, tgs.tag_name
FROM BooksDB.dbo.book_tags AS bt
INNER JOIN BooksDB.dbo.books AS BKS ON bt.goodreads_book_id = bks.book_id
INNER JOIN BooksDB.dbo.tags AS tgs ON bt.tag_id = tgs.tag_id
WHERE tgs.tag_name LIKE 'female_author%' OR tgs.tag_name LIKE 'woman-author%'
--I got better results looking for popular relevant tags and then filtering for them than I did by using GROUP BY or HAVING
```

### Event 2: Choose another event from March/Spring

Write a query to return authors, titles, ratings, and `tag_id` that you would want to promote during your chosen event.

```python
-- Solution
-- I chose Easter
SELECT bks.authors, bks.title, bks.average_rating, bt.tag_id, bt.goodreads_book_id, tgs.tag_id, tgs.tag_name
FROM BooksDB.dbo.book_tags AS bt
INNER JOIN BooksDB.dbo.books as bks ON bt.goodreads_book_id = bks.book_id
INNER JOIN BooksDB.dbo.tags AS tgs ON bt.tag_id = tgs.tag_id
Where tgs.tag_name LIKE 'easter' OR tgs.tag_name LIKE 'spring'
```

# Part 3: Summarize your Work

For each event write at least one query that joins any two tables in `BooksDB` to support your choice and record you thoughts as to why you used the paticlular query. At least one of your queries needs to include a `HAVING` clause.

```python
-- Event 1 Query
-- Mothers' Day
SELECT bks.book_id, bks.title, bks.average_rating, tgs.tag_id, tgs.tag_name
FROM BooksDB.dbo.book_tags AS bt
INNER JOIN BooksDB.dbo.books AS bks ON bt.goodreads_book_id = bks.book_id
INNER JOIN BooksDB.dbo.tags AS tgs ON bt.tag_id = tgs.tag_id
WHERE tag_name LIKE 'mom' OR tag_name LIKE 'mother%' OR tag_name LIKE 'parenting'
GROUP BY bks.title, bks.book_id, bks.average_rating, tgs.tag_name, tgs.tag_id
HAVING average_rating > 2.5
```

### Summarize Event 1

Looking for books for mothers or about motherhood

```python
-- Event 2 Query
-- Memorial Day
SELECT bks.book_id, bks.title, bks.average_rating, tgs.tag_id, tgs.tag_name
FROM BooksDB.dbo.book_tags AS bt
INNER JOIN BooksDB.dbo.books AS bks ON bt.goodreads_book_id = bks.book_id
INNER JOIN BooksDB.dbo.tags AS tgs ON bt.tag_id = tgs.tag_id
WHERE tag_name LIKE 'soldier' OR tag_name LIKE 'veteran' OR tag_name LIKE 'american History'
GROUP BY bks.title, bks.book_id, bks.average_rating, tgs.tag_name, tgs.tag_id
HAVING average_rating > 2.5
```
